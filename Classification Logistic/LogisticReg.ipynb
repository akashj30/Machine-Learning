### Logistic Regression - This is used for classification only

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

### Step 1 : Read the train data


```python
import pandas as pd
df = pd.read_csv("iris.csv")
df.head()
```

### Step 2: Performing basic data quality checks


```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

### Step 3: Seperate X and Y(species)


```python
X = df.drop(columns=['species'])
Y = df[['species']]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

### Step 4: Preprocess data for X


```python
X.dtypes
```

```python

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute',SimpleImputer(strategy= 'mean')),
                           ('scaler',StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

### Step 5: Train test


```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre,Y, test_size=0.33,random_state=21)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Step 5: Model Building

```python

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
model.classes_
```

```python
model.intercept_
```

```python
xtrain.columns
```

```python
model.coef_
```

### Step 6: Model Evaluation


```python
# Accuracy in train
model.score(xtrain,ytrain)
```

```python
# Accuracy in test
model.score(xtest,ytest)
```

```python
ypred_train = model.predict(xtrain)
ypred_test = model.predict(xtest)
```

```python
ypred_train[0:5]
```

```python
ytrain.head()
```

```python
ypred_test[0:5]
```

```python
ytest.head()
```

```python
model.classes_
```

```python
ypred_test_prob = model.predict_proba(xtest)
```

```python
ypred_test_prob[0:5]
```

```python
# Confusion Matrix
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(model,xtest,ytest)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred_test))
```

### Because f1 score for above in 0.94 in test it is a good model


```python
from sklearn.model_selection import  cross_val_score
scores = cross_val_score(model, xtrain, ytrain, cv = 5,scoring='f1_macro')
scores
```

```python
scores.mean()
```

### Above is good model i can use it for out of sample prediction
