### Step 1: Read the dataset


```python
import pandas as pd
url = "https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/main/AirPassengers.csv"
df = pd.read_csv(url, parse_dates=True, index_col="Month")
df.head()
```

### Step 2- Perform basic data quality checks


```python
df.info()
```

```python
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot(df["Passengers"])
plt.xlabel("Months")
plt.ylabel("No. of Passengers")
plt.title("Air Passengers over time")
plt.grid()
plt.show()
```

### Step 3 - Build the ARIMA model


```python
from pmdarima.arima import auto_arima
```

```python
model = auto_arima(df["Passengers"], m=12, seasonal=True)
```

```python
model.order
```

### Based on above model , estimate the air pasenger demand for next four years

```python
forecast, conf = model.predict(n_periods=48, return_conf_int=True)
```

```python
forecast
```

```python

df_forecast = forecast.to_frame()
df_forecast.columns = ["Forecast"]
df_forecast
```

```python
conf
```

```python
df_forecast[["lower", "upper"]] = conf
```

```python
df_forecast
```

```python
df_forecast = df_forecast.apply(round)
df_forecast
```

```python
df_forecast.to_csv("Forecast.csv")
```

### Visualize results


```python
plt.figure(figsize=(10, 5))
plt.plot(df["Passengers"], label="Actual")
plt.plot(df_forecast["Forecast"], label="Forecast")
plt.fill_between(
    df_forecast.index,
    df_forecast["lower"],
    df_forecast["upper"],
    color="gray",
    alpha=0.5
)
plt.xlabel("Months")
plt.ylabel("No. of Passengers")
plt.title("Air Passengers over time")
plt.legend(loc="upper left")
plt.grid()
plt.show()
```
