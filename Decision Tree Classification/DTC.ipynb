### Decision Tree Classification
Decision Tree is also called as CART - Classification and Regression Tree

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
import pandas as pd
df = pd.read_csv("iris.csv")
df.head()
```

### Step 2: Perform Basic Data quality checks


```python
df.info()
```

### 

### Step 3: Seperate X and Y(species)


```python
X = df.drop(columns=['species'])
Y = df[['species']]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

```python
Y.value_counts().plot(kind='bar')
```

### Step 4: Apply Preprocessing on X


```python
X.dtypes
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = make_pipeline(SimpleImputer(strategy='mean'),
                         StandardScaler()).set_output(transform = 'pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

### Step 5: Apply Train Test split


```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X_pre,Y,test_size=0.33,random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Step 6: Model Building


```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(
    max_depth = 1,
    min_samples_split= 5,
    min_samples_leaf= 5,
    criterion="entropy"
)
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

### Hyperparameter tuning


```python
params = {
    "max_depth" : [1,2,3,4,5,6,7,8,9],
    "min_samples_split" : [5,6,7,8,9],
    "min_samples_leaf" : [5,6,7,8,9],
    "criterion" : ["gini","entropy"]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
dtc = DecisionTreeClassifier()
gscv = GridSearchCV(dtc, param_grid=params, cv=5, scoring='f1_macro')
gscv.fit(xtrain, ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_dtc = gscv.best_estimator_
best_dtc
```

### Model Evaluation


```python
best_dtc.score(xtrain,ytrain)
```

```python
best_dtc.score(xtest,ytest)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(best_dtc,xtrain,ytrain)
```

```python
ConfusionMatrixDisplay.from_estimator(best_dtc,xtest,ytest)
```

```python
ypred_train = best_dtc.predict(xtrain)
ypred_test = best_dtc.predict(xtest)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred_test))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(best_dtc,xtrain,ytrain,cv=5,scoring='f1_macro')
scores
```
