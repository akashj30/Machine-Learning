### Step 3: Seperate X and Y(Survived)
1. PassengerId, Name, Ticket are statistically insignificant

2. Cabin is removed because it has lots of missing values

```python
X = df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin' ,'Survived'])
Y = df[['Survived']]
```

```python
X.head()
```

```python
Y.head()
```

### Step 4: Preprocessing on X


```python
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy='mean'),
    StandardScaler()
)
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy='most_frequent'),
    OneHotEncoder(handle_unknown='ignore',sparse_output=False)
)
```

```python
pre = ColumnTransformer([('num',num_pipe,con),
                         ('cat',cat_pipe,cat)]).set_output(transform='pandas')
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

### Step 5: Apply Train Test Split


```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X_pre,Y,test_size=0.2,random_state=21)
```

```python
ytrain.head()
```

```python
ytest.head()
```

```python
xtest.shape
```

### Step 6: Model Building


```python
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier(
    n_estimators=1,
    max_depth=1,
    min_samples_split=5,
    min_samples_leaf=5,
    criterion='entropy'
)
model1.fit(xtrain, ytrain)
```

```python
model1.score(xtrain, ytrain)
```

```python
model1.score(xtest, ytest)
```

### Hyperparameter tuning Random Forest


```python
params = {
    "n_estimators" : [10, 50, 100, 200, 300],
    "max_depth" : [2, 3, 4, 5],
    "min_samples_split" : [5, 6, 7, 8, 9, 10]
}
```

### Use randomized Search CV for faster tuning


```python
from sklearn.model_selection import RandomizedSearchCV
rf = RandomForestClassifier()
rscv_rf = RandomizedSearchCV(rf, param_distributions=params, cv=5, scoring='f1_macro')
rscv_rf.fit(xtrain, ytrain)
```

```python
rscv_rf.best_params_
```

```python
rscv_rf.best_score_
```

```python
best_rf = rscv_rf.best_estimator_
best_rf
```
